```python
class Node:
    def __init__(self, val=0, next=None):
        self.data = val
        self.next = next
```

```python
def create_LL_from_List(l):
    if not l:
        return None  # ! handle empty list

    head = Node(l[0])
    temp = head

    for i in l[1:]:
        temp.next = Node(i)
        temp = temp.next

    return head
```

```python
def print_LL(head):
    
    temp=head

    while(temp!=None):

        print(temp.data,end="->")
        temp=temp.next
```

```python
l1 = create_LL_from_List([3,7])
l2 = create_LL_from_List([9,2])
print_LL(l1)
print()
print_LL(l2)
```

```python
def add_two_number(l1,l2):
    
    dummy=Node(0)
    tail=dummy
    add=0

    while l1 and l2:
        
        tot=l1.data+l2.data+add
        
        r=tot%10
        
        newNode=Node(r)
        tail.next=newNode
        tail=newNode
        add=tot//10
        
        l1=l1.next
        l2=l2.next
        
    
    if l2 is None:
        
        while l1:
            tot=l1.data+add
            r=tot%10
            newNode=Node(r)
            tail.next=newNode
            tail=newNode
            add=tot//10
            l1=l1.next
            
    if l1 is None:

        while l2:
            tot = l2.data+add
            r = tot % 10
            newNode = Node(r)
            tail.next = newNode
            tail = newNode
            add = tot//10
            l2 = l2.next
            
            
    if add==1:
        newNode = Node(add)
        tail.next = newNode
        tail = newNode
        

    return dummy.next

    
```

```python
head=add_two_number(l1, l2)
print_LL(head)
```
